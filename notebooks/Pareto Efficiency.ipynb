```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import Plotting_settings
import itertools as it
```

```python
rubbers = pd.read_csv('data/Rubbers_new.csv') 
blades = pd.read_csv('data/Blades_new.csv')
print('Rubbers')
print(rubbers.head())
print('\n')
print('Blades')
print(blades.head())
```

```python
#All the brands that make rubbers
brands = ["Butterfly", "XIOM", "DHS", "Tibhar", "Andro", "JOOLA", "Killerspin", "Yasaka",
          "Donic", "Stiga", "RITC", "LKT", "Gambler", "Sanwei", "GEWO", "TSP", "Victas"]

#New Data Frame that will contain all the averages for each feature and each brand
new_rubbers = pd.DataFrame(index = brands, columns= ['Price_avg', 'Spin_avg', 'Speed_avg', 'Control_avg'])

#Create a bar chart with average price, spin, speed, and control for each brand
def combine_brands(data):
    for brand in brands:
        key_values = data['Product_Name'].apply(lambda x: x.find(brand) != -1).sort_values(ascending = False)
        indices = key_values.index[key_values.values == True]
        frame = data.iloc[indices].sort_index()
        new_rubbers.loc[brand,:] = [frame['Price'].mean(), frame['Spin'].mean(),
                                                   frame['Speed'].mean(), frame['Control'].mean()]      
    return new_rubbers

combine_brands(rubbers)
new_rubbers = new_rubbers.astype(int)
```

```python
#Plot a Heatmap of the rubber stats for each brand
fig, ax = plt.subplots(1,1,figsize = (10,10))
cols = ["Price_avg", "Spin_avg", "Speed_avg", "Control_avg"]
sns.heatmap(new_rubbers[cols], annot = True, ax=ax, linewidth = 1, fmt = '.3g', cmap = 'Blues')
```

There exists a clear trade-off between Average Speed rating and Average Control rating

```python
brands = ["XIOM", "Andro", "JOOLA", "Butterfly", "Yasaka", "Donic",  "Stiga", "Gambler", "Cornilleau",
          "Sanwei", "GEWO", "TSP", "Victas"]

colors = ["#fc5c65", "#eb3b5a", "#fd9644", "#fa8231", "#fed330", "#f7b731", "#4b7bec", "#3867d6","#45aaf2", "#2d98da", "#a55eea",
              "#8854d0", "#26de81", "#20bf6b", "#d1d8e0", "#a5b1c2", "#2bcbba", "#0fb9b1", "#778ca3", "#4b6584"]

new_blades = pd.DataFrame(index = brands, columns= ['Price_avg', 'Speed_avg', 'Control_avg', 'Weight_avg'])

#Create a bar chart with average price, spin, speed, and control for each brand
def combine_brands(data):
    for brand in brands:
        key_values = data['Product_Name'].apply(lambda x: x.find(brand) != -1).sort_values(ascending = False)
        indices = key_values.index[key_values.values == True]
        frame = data.iloc[indices].sort_index()
        new_blades.loc[brand,:] = [frame['Price'].mean(), frame['Speed'].mean(),
                                                   frame['Control'].mean(), frame['Weight (g)'].mean()]      
    return new_blades

combine_brands(blades)
new_blades = new_blades.astype(int)
```

```python
#Plot a Heatmap of the rubber stats for each brand
fig, ax = plt.subplots(1,1,figsize = (10,10))
cols = ["Price_avg", "Speed_avg", "Control_avg", "Weight_avg"]
sns.heatmap(new_blades[cols], annot = True, ax=ax, linewidth = 1, fmt = '.3g', cmap = 'Greens')
```

Once again the trade-off between Blade Speed rating and Blade Control rating appears.

```python
#Creating a new dataframe where the the product name is the index
#This will be used later in the Pareto Function
rubbers_options = rubbers.set_index(rubbers['Product_Name']).drop(columns = 'Product_Name')
blades_options = blades.set_index(blades['Product_Name']).drop(columns = 'Product_Name')
print("Rubbers:")
print(rubbers_options.head())
print("\nBlades:")
print(blades_options.head())
```

```python
#Renaming the "Product_Name" column to Rubber and Blade for each product
rubbers = rubbers.rename(columns = {"Product_Name": 'Rubber'})
blades = blades.rename(columns = {'Product_Name': 'Blade'})
print(rubbers.head())
print(blades.head())
```

```python
#Finding the optimal combination of speed and control
def check(rubber_name, blade_name):
    rubber = rubbers[rubbers['Rubber'] == rubber_name]
    blade = blades[blades['Blade'] == blade_name]
    
    #Summing the Rubber Values and the Blade Values
    stats = pd.concat([rubber[['Speed', 'Control', 'Spin', 'Price']], blade[['Speed', 'Control', 'Weight (g)', 'Price']]]).sum()
    
    #Creating a hierarchical index: first the rubber name and then the blade name
    index = pd.MultiIndex.from_tuples([(rubber_name, blade_name)], names = ['Rubber', 'Blade'])
    
    #Data Frame with the values for the particular created combination
    df =pd.DataFrame(stats).transpose()
    df.index = index
    return df
    
#List of tuples for every configuration
config_all = it.product(rubbers_options.index, blades_options.index)
    
#Data frame that will store the stats for each configuration
config_base = pd.DataFrame()

#For each combination create a dataframe for that combination and add it to the config_base dataframe
for (r,b) in config_all:
        this_config = check(r,b)
        config_base = config_base.append(this_config)
```

```python
print(config_base.head())
print("\nThe number of configurations is: {}".format(len(config_base)))
```

```python
#Finding if combination is at the pareto frontier

def is_pareto_front(row, xlabel, ylabel):
    x = row[xlabel]
    y = row[ylabel]
    
    #Look for points with the same y value but larger x value
    is_max_x = config_base.loc[config_base[ylabel] == y].max()[xlabel]<=x
    #Look for points with the same x value but larger y value
    is_max_Y = config_base.loc[config_base[xlabel] == x].max()[ylabel]<=y
    #Look for points that are larger than both x and y
    is_double = len(config_base.loc[(config_base[xlabel]>x) & (config_base[ylabel]>y)]) == 0
    
    return is_max_x and is_max_Y and is_double

#From the dataframe with all configurations check if each configuration is on the pareto frontier
is_pareto = config_base.apply(lambda row: is_pareto_front(row, 'Speed', 'Control'), axis = 1)

#Store all the values that are on the pareto frontier in a new dataframe
config_pareto = config_base.loc[is_pareto].sort_values('Speed')
```

```python
#Creating a new dataframe that doesn't contains all the values on the Pareto Frontier
base_no_pareto = pd.concat([config_base, config_pareto]).drop_duplicates(keep=False)
base_no_pareto.head()
```

```python
#Plotting all configurations
fig, ax = plt.subplots(1,1,figsize = (30,15))
sns.regplot(x = 'Speed', y = 'Control', data = base_no_pareto, fit_reg = False,
           ax =ax, color = '#26de81')

#Pareto Points
sns.regplot(x = 'Speed', y = 'Control', data = config_pareto, fit_reg = False,
           ax =ax, color = '#a55eea', scatter_kws={"color":"#a55eea","alpha":0.8,"s":100})

#Plot Pareto Frontier
plt.plot(config_pareto['Speed'], config_pareto['Control'], '--',
         label = 'Pareto Frontier', alpha = 4.0, color = "#3867d6")

#More plotting settings
plt.legend(loc = 'upper right', fontsize = 25)
plt.title("Control vs Speed Combinations for Ping Pong Rubbers and Blades", fontsize = 35)
plt.xlabel('Speed', fontsize=25)
plt.ylabel('Control', fontsize=25)
```

```python
print('Total Possible Combinations: {}'.format(len(config_base)))
print("Efficient Combinations that lay on the Pareto Frontier: {}".format(len(config_pareto)))
```
