```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Plotting_settings
from sklearn.cluster import KMeans
```

```python
rubber = pd.read_csv("data/Rubbers_new.csv")
blade = pd.read_csv("data/Blades_new.csv")

print(rubber.head())
print("\n")
print(blade.head())
```

```python
#Creating KMeans scatterplot for Rubbers
rubber_X = rubber[['Speed', 'Spin']].values
rubber_y = rubber[['Product_Name']]

kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans.fit(rubber_X)
y_kmeans = kmeans.predict(rubber_X)

import matplotlib.colors
cmap = matplotlib.colors.ListedColormap(["#2d98da", "#20bf6b","#0fb9b1"])

plt.scatter(rubber_X[:, 0], rubber_X[:, 1], c=y_kmeans, s=50, cmap=cmap)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("Speed")
plt.ylabel("Spin")
plt.title("KMeans Clustering for Rubbers")
plt.show()

#Creating KMeans scatterplot for Blades
blade_X = blade[['Speed', 'Control']].values
blade_y = blade[['Product_Name']]

clf = KMeans(n_clusters = 3, random_state = 0)
clf.fit(blade_X)
y_clf = clf.predict(blade_X)

plt.scatter(blade_X[:,0], blade_X[:, 1], c= y_clf, s = 50, cmap = cmap)
centers = clf.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c = 'black', s = 200, alpha = 0.5)
plt.xlabel("Speed")
plt.ylabel("Control")
plt.title("KMeans Clustering for Blades")
plt.show()
```
